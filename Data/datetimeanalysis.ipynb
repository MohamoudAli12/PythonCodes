```python
import pandas as pd
df = pd.read_csv('CRYPTOCURRENCY_US_KRAKEN_ETHUSD.csv')
df
#df.loc[0, 'Date'].day_name()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df
```

```python
df.loc[10, 'Date'].day_name()
```

```python
df['DayofWeek'] = df['Date'].dt.day_name()
df
```

```python
df['Date'].max()
df['Date'].min()
df['Date'].max() - df['Date'].min()
```

```python
filt = (df['Date'] >= '2022-04-01') & (df['Date'] < '2022-08-01') 
df.loc[filt]
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df['Close'] = df['Close'].apply(lambda x: pd.to_numeric(x.replace(',', '')))
df['Low'] = df['Low'].apply(lambda x: pd.to_numeric(x.replace(',', '')))
df['Open'] = df['Open'].apply(lambda x: pd.to_numeric(x.replace(',', '')))
df['High'] = df['High'].apply(lambda x: pd.to_numeric(x.replace(',', '')))
```
